```python
labels=['real','HF','LF','invert','picassoL','picassoR','beak','righteye','lefteye', 'eyes','topheavy']
all_c = pd.DataFrame(0, index=labels, columns=labels)
label_grouping = [['real'],['HF','LF'],['invert'],['picassoL','picassoR'],['beak'],['righteye','lefteye'],['eyes'],['topheavy']]
for group in label_grouping:
    data = SVMdataset(training_data = 'faceChicks_wide.csv', labels_index = 1, trials_axis = 0, normalization_type = 'normalized', normalization_ref = 'all')
    AdjustSplit.move_categories_to_test(data, group) ### adjust to prevent error if this function is called after spli!!!!!!!!!!
    AdjustSplit.split_dataset(data, split_type = 'perc', validation_percentage = 0.1, n_combinations = 5)
    RunModel.runSVM(data, shuffle=False, verbose=True)
    cms = RunModel.printConfusionMatrix(data, ['training', 'validation','test'], order_labels = labels)
    all_c += cms[2].fillna(0)

sns.heatmap(all_c, annot=True, fmt=".0%")
```
